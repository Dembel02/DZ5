Заполним пропуски в числовых столбцах с возрастом и номер кабины (логично было бы заполнить номер кабины по
имеющим номерам кабин по данной фамилии имеющих родственников, SibSp числобратьев, сестер или супругов на борту
у человека и Parch количество родителей или детей, с которыми путешествовал каждый пассажир), в excel я бы разбил по столбцам столбец Name и искал там родственников, будем думать что датасет составлен так что родственники идут друг за другом и для заполнения номера кабины используем функцию 

```python
df['Cabin'] = df['Cabin'].fillna(method='bfill')
df
```

```python
df['Cabin'] = df['Cabin'].fillna(method='ffill')
df
```

```python
median_age = df.Age.median()
df.fillna({'Age' : median_age}, inplace = True)
df
```

```python
# для практики можно удалить 2 строки Embarked Порт посадки 
df.dropna(axis = 'index', subset = ['Embarked'], inplace = True)
df
```

```python
#Выводим количество уникальных значений по столбцам
df.nunique()
```

```python
# Выводим данные по максимальным минимальным значениям числовых столбцов
df.describe()
```

```python
df_part_class = df[['Pclass', 'Survived']]
fuel_type_rate = df_part_class.groupby('Pclass').apply(lambda x: (len(x[x['Survived'] == 1])/len(x), len(x[x['Survived'] == 1])))


for i, rate in fuel_type_rate.items():
  print(f'с классом круиза {i} доля выживших: {rate[0]}, общее число : {rate[1]}')
```

вывод: чем выше класс билета пасажира, тем больше выживших

```python
df_part_age = df[['Age', 'Survived']]
fuel_type_rate = df_part_age.groupby('Age').apply(lambda x: (len(x[x['Survived'] == 1])/len(x), len(x[x['Survived'] == 1])))


for i, rate in fuel_type_rate.items():
  print(f'с возрастом {i} доля выживших: {rate[0]}, общее число : {rate[1]}')
```

Вывод: младенцы с возрастом до 2 лет большая доля выживших, далее кореляция по возрасту проподает

```python
df_part_sex = df[['Sex', 'Survived']]
fuel_type_rate = df_part_sex.groupby('Sex').apply(lambda x: (len(x[x['Survived'] == 1])/len(x), len(x[x['Survived'] == 1])))


for i, rate in fuel_type_rate.items():
  print(f'с полом {i} доля выживших: {rate[0]}, общее число : {rate[1]}')
```

Вывод: доля выживших женщин больше чем мужчин.
